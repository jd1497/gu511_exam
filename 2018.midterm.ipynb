```py
HEADERS = ['x', 'y', 'z']

def main(csv_path, pkl_path):
    # read a csv file
    fobj_csv = open(csv_path, 'r')
    
    records = []
    for line in fobj_csv:
        records.append(
            dict(zip(HEADERS, records.strip().split(',')))
        )
    
    fobj_csv.close()
    
    # save this as a pickle object
    fobj_pkl = open(pkl_path, 'w')
    
    pickle.dump(records, fobj_pkl)
    
    fobj_pkl.close()


main('data.csv', 'data.pkl')
```
